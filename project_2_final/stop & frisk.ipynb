```python
#Seeing how many people were stop and frisked in each racial group
df_clean['SUSPECT_RACE_DESCRIPTION'].value_counts()
```

```python
df_2 = df_clean[['STOP_LOCATION_PRECINCT','SUSPECT_RACE_DESCRIPTION']]
```

```python
df_race = df_2.groupby(['STOP_LOCATION_PRECINCT','SUSPECT_RACE_DESCRIPTION']).value_counts().reset_index(name= 'Count')
```
